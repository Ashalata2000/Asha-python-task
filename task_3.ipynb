 Instance Method


Q1. Write a program to display a student's name using an instance method.

```python
class Student:
    def __init__(self, name):
        self.name = name

    def display(self):
        print("Name:", self.name)

s = Student("Rahul")
s.display()
```

Q2. Write a program to calculate the area of a rectangle using an instance method.

```python
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        print("Area =", self.length * self.width)

r = Rectangle(10, 5)
r.area()
```

Q3. Write a program to print employee details using an instance method.

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def details(self):
        print(self.name, self.salary)

e = Employee("Amit", 50000)
e.details()
```

Q4. Write a program to check whether a number is even or odd using an instance method.

```python
class Number:
    def __init__(self, num):
        self.num = num

    def check(self):
        if self.num % 2 == 0:
            print("Even")
        else:
            print("Odd")

n = Number(17)
n.check()
```

Q5. Write a program to display a car's information using an instance method.

```python
class Car:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def show(self):
        print(self.brand, self.model)

c = Car("Toyota", "Innova")
c.show()
```

Q6. Write a program to calculate the sum of two numbers using an instance method.

```python
class Sum:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    def add(self):
        print("Sum =", self.a + self.b)

s = Sum(20, 30)
s.add()
```

2. Class Method

Q1. Write a program to display the school name using a class method.

```python
class Student:
    school = "ABC School"

    @classmethod
    def show_school(cls):
        print(cls.school)

Student.show_school()
```

Q2. Write a program to update a company name using a class method.

```python
class Employee:
    company = "ABC Ltd"

    @classmethod
    def change_company(cls, name):
        cls.company = name

Employee.change_company("XYZ Ltd")
print(Employee.company)
```

Q3. Write a program to count the number of objects using a class method.

```python
class Student:
    count = 0

    def __init__(self):
        Student.count += 1

    @classmethod
    def total(cls):
        print(cls.count)

Student()
Student()
Student()
Student.total()
```

Q4. Write a program to display a class variable using a class method.

```python
class College:
    name = "National College"

    @classmethod
    def display(cls):
        print(cls.name)

College.display()
```

Q5. Write a program to change the course name using a class method.

```python
class Course:
    course = "Python"

    @classmethod
    def update(cls, new_course):
        cls.course = new_course

Course.update("Java")
print(Course.course)
```

Q6. Write a program to create an object using a class method

```python
class Student:
    def __init__(self, name):
        self.name = name

    @classmethod
    def create(cls):
        return cls("Ravi")

s = Student.create()
print(s.name)
```

3. Static Method

Q1. Write a program to add two numbers using a static method.

```python
class Calculator:
    @staticmethod
    def add(a, b):
        print(a + b)

Calculator.add(10, 20)
```

Q2. Write a program to find the square of a number using a static method.

```python
class Math:
    @staticmethod
    def square(n):
        print(n * n)

Math.square(6)
```

Q3. Write a program to check whether a number is positive or negative using a static method

```python
class Check:
    @staticmethod
    def sign(num):
        if num >= 0:
            print("Positive")
        else:
            print("Negative")

Check.sign(-8)
```

Q4. Write a program to convert Celsius to Fahrenheit using a static method.

```python
class Temperature:
    @staticmethod
    def convert(c):
        print((c * 9/5) + 32)

Temperature.convert(25)
```

Q5. Write a program to calculate the factorial of a number using a static method.

```python
class Factorial:
    @staticmethod
    def fact(n):
        result = 1
        for i in range(1, n + 1):
            result *= i
        print(result)

Factorial.fact(5)
```

Q6. Write a program to check whether a year is a leap year using a static method

```python
class LeapYear:
    @staticmethod
    def check(year):
        if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
            print("Leap Year")
        else:
            print("Not Leap Year")

LeapYear.check(2024)
```
